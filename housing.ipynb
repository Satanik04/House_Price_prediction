```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("housing.csv")
```

```python
df.head()
```

1. longitude: A measure of how far west a house is; a higher value is farther west
2. latitude: A measure of how far north a house is; a higher value is farther north
3. housingMedianAge: Median age of a house within a block; a lower number is a newer building
4. totalRooms: Total number of rooms within a block
5. totalBedrooms: Total number of bedrooms within a block
6. population: Total number of people residing within a block
7. households: Total number of households, a group of people residing within a home unit, for a block
8. medianIncome: Median income for households within a block of houses (measured in tens of thousands of US Dollars)
9. medianHouseValue: Median house value for households within a block (measured in US Dollars)
10. oceanProximity: Location of the house w.r.t ocean/sea

```python
df["total_bedrooms"]=df["total_bedrooms"].fillna(df["total_bedrooms"].median())
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
```

```python
encoded_data=encoder.fit_transform(df[['ocean_proximity']])
```

```python
encoded_data=encoded_data.toarray()
```

```python
df_encoder=pd.DataFrame(
    encoded_data,
    columns=encoder.get_feature_names_out(['ocean_proximity'])
)
```

```python
df_encoder
```

```python
df=pd.concat([df,df_encoder],axis=1)
```

```python
df.drop('ocean_proximity',axis=1)
```

```python
#df_encoder.reset_index(drop=True,inplace=True)
#df.reset_index(drop=True,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X=df.drop("median_house_value",axis=1)
y=df["median_house_value"]
```

```python
X.drop("ocean_proximity",axis=1,inplace=True)
```

```python
y
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print("R2 score",r2_score(y_test,y_pred))
print("mean absolute error",mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
decisiontree=DecisionTreeRegressor()
```

```python
decisiontree.fit(X_train,y_train)
```

```python
y_predict=decisiontree.predict(X_test)
```

```python
print("R2_score is:",r2_score(y_test,y_predict))
```

```python
from sklearn.ensemble import RandomForestRegressor
Rf=RandomForestRegressor()
```

```python
Rf.fit(X_train,y_train)
```

```python
y_pred=Rf.predict(X_test)
```

```python
print("R2 score is:",r2_score(y_test,y_pred))
```

```python
Rf=RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42
)
```

```python
Rf.fit(X_train,y_train)
```

```python
y_pred=Rf.predict(X_test)
```

```python
print("r2 score:",r2_score(y_test,y_pred))
```
